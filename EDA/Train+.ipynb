The stats shown are:
    <ul style="margin-top:0">
    <li><strong>Count</strong>: number of records per class that exist in the dataset</li>

    </ul>

```python
dosClasses = {}
probeClasses = {}
r2lClasses = {}
u2rClasses = {}

for classes in dos_types:
    dosClasses[classes] = dos[dos["Attack_Type(42)"].isin([classes])]

for classes in probe_types:
    probeClasses[classes] = probe[probe["Attack_Type(42)"].isin([classes])]

for classes in r2l_types:
    r2lClasses[classes] = r2l[r2l["Attack_Type(42)"].isin([classes])]

for classes in u2r_types:
    u2rClasses[classes] = u2r[u2r["Attack_Type(42)"].isin([classes])]
```

#### DoS vs Probe vs U2R vs R2L Feature Count Distribution

<br>
The stats shown are:
    <ul style="margin-top:0">
    <li><strong>Count</strong>: number of records per class that exist in the dataset</li>
        <li><strong>Subset Percentage</strong>: percentage of records this class makes up in the subset</li>
        <li><strong>Dataset Percentage</strong>: percentage of records this class makes up in the entire dataset</li>
    </ul>

```python
dosClasses = {}
probeClasses = {}
r2lClasses = {}
u2rClasses = {}

for classes in dos_types:
    dosClasses[classes] = dos[dos["Attack_Type(42)"].isin([classes])]

for classes in probe_types:
    probeClasses[classes] = probe[probe["Attack_Type(42)"].isin([classes])]

for classes in r2l_types:
    r2lClasses[classes] = r2l[r2l["Attack_Type(42)"].isin([classes])]

for classes in u2r_types:
    u2rClasses[classes] = u2r[u2r["Attack_Type(42)"].isin([classes])]
```

```python
dosClassesStats = {
    "count": {},
    "percent":{
        "dos":{}, 
        "total":{}
    }}
probeClassesStats = {
    "count":{},
    "percent":{
        "probe":{}, 
        "total":{}
    }}
r2lClassesStats = {
    "count":{}, 
    "percent":{
        "r2l":{}, 
        "total":{}
    }}
u2rClassesStats = {
    "count":{}, 
    "percent":{
        "u2r":{}, 
        "total":{}
    }}
normalClassesStats = {
    "total":{
        "count":{}, 
        "percent":{}
    }}
for classes in dos_types:
    dosClassesStats["count"][classes] = dosClasses[classes].shape[0]
    dosClassesStats["percent"]["dos"][classes] = round(dosClassesStats["count"][classes]/dos.shape[0], 4)
    dosClassesStats["percent"]["total"][classes] = round(dosClassesStats["count"][classes]/totalRecords, 4)

for classes in probe_types:
    probeClassesStats["count"][classes] = probeClasses[classes].shape[0]
    probeClassesStats["percent"]["probe"][classes] = round(probeClassesStats["count"][classes]/probe.shape[0], 4)
    probeClassesStats["percent"]["total"][classes] = round(probeClassesStats["count"][classes]/totalRecords, 4)
    
for classes in r2l_types:
    r2lClassesStats["count"][classes] = r2lClasses[classes].shape[0]
    r2lClassesStats["percent"]["r2l"][classes] = round(r2lClassesStats["count"][classes]/r2l.shape[0], 4)
    r2lClassesStats["percent"]["total"][classes] = round(r2lClassesStats["count"][classes]/totalRecords, 4)
    
for classes in u2r_types:
    u2rClassesStats["count"][classes] = u2rClasses[classes].shape[0]
    u2rClassesStats["percent"]["u2r"][classes] = round(u2rClassesStats["count"][classes]/u2r.shape[0], 4)
    u2rClassesStats["percent"]["total"][classes] = round(u2rClassesStats["count"][classes]/totalRecords, 4)
                                                         
```
